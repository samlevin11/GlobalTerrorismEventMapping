```python
import pandas as pd
import numpy as np
```

```python
inCSV = r"PATH\globalterrorismdb_0718dist.csv"
```

```python
allEvents = pd.read_csv(inCSV, encoding='ANSI')
```

```python
realtedEvents = allEvents[allEvents['related'].notna()]
```

```python
len(allEvents['gname'].unique())
```

```python
gKnownEvents = allEvents[allEvents['gname'] != 'Unknown']
```

```python
gKnownEvents
```

```python
gKnownEvents.to_csv(r"PATH\globalterrorismdb_0718dist_gKnown.csv")
```

```python
grouped = gKnownEvents.groupby(['gname'])
```

```python
grouped.head()
```

```python
top10 = grouped['eventid'].count().sort_values(ascending=False).head(10)
```

```python
top10
```

```python
top10names = pd.DataFrame(top10).index.values.tolist()
```

```python
type(top10names)
```

```python
top10names
```

```python
top10groupsEvents = allEvents[allEvents['gname'].isin(top10names)]
```
